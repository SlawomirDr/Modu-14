```python
import pandas as pd
movies = pd.read_csv('tmdb_movies.csv')
movies
```

```python
genres = pd.read_csv('tmdb_genres.csv')
genres
```

```python
movies['vote_count'].quantile(q=0.75)
```

```python
movies[movies['vote_count'] > 737].sort_values('vote_average', ascending=False).head(10)
```

```python
movies['release_date'] = pd.to_datetime(movies['release_date'])
money = movies[(movies['release_date'] > '2010/01/01') & (movies['release_date'] < '2016/12/31')].groupby(pd.Grouper(key='release_date',freq='Y')).mean()
money
```

```python
import matplotlib.pyplot as plt
import numpy as np
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
def million(x, pos):
        return 'PLN {:2.1f}M'.format(x*1e-6)
x = money.index
y = money['budget']
z = money['revenue']
axes.set_title('Średni przychód i budżety filmów w latach 2010-2016')
#money['revenue'].plot(kind='bar')
axes.plot(x,y, color="red", label='budget')
axes.plot(x,z, color="blue", alpha=0.5, label='revenue')
formatter = plt.FuncFormatter(million)
axes.yaxis.set_major_formatter(formatter)
axes.legend(loc=(1.05,0.8))
plt.show()
```

```python
complete_table.groupby('genres').mean('runtime').sort_values('runtime', ascending=False)
```
